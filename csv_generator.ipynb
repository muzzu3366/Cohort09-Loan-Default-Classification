```python
import pandas as pd
import numpy as np
```

```python


num_records = 5000


applicant_id = np.arange(100000, 100000 + num_records)


age = np.random.randint(21, 65, num_records)


gender = np.random.choice(
    ["Male", "Female"],
    num_records
)


regions = [
    "South Asia",
    "Sub-Saharan Africa",
    "Latin America",
    "Southeast Asia"
]

region = np.random.choice(regions, num_records)


education = np.random.choice(
    ["Primary", "Secondary", "Graduate", "Postgraduate"],
    num_records
)


employment = np.random.choice(
    ["Farmer", "Small Business", "Self-employed", "Salaried"],
    num_records
)


employment_years = np.random.randint(0, 30, num_records)


monthly_income = np.random.randint(6000, 50000, num_records)


loan_amount = np.random.randint(5000, 150000, num_records)


loan_duration = np.random.choice(
    [6, 12, 18, 24, 36],
    num_records
)


interest_rate = np.round(
    np.random.uniform(12, 28, num_records),
    2
)


loan_purpose = np.random.choice(
    ["Business", "Education", "Agriculture", "Housing", "Medical"],
    num_records
)


credit_score = np.random.randint(300, 850, num_records)


dependents = np.random.randint(0, 6, num_records)


previous_loans = np.random.randint(0, 6, num_records)


repayment_history = np.random.choice(
    ["Good", "Average", "Poor"],
    num_records,
    p=[0.5, 0.3, 0.2]  # probability distribution
)


collateral = np.random.choice(
    ["Yes", "No"],
    num_records,
    p=[0.4, 0.6]
)


annual_income = monthly_income * 12
debt_to_income_ratio = np.round(
    loan_amount / annual_income,
    3
)



loan_status = []

for i in range(num_records):

    risk = 0

    if debt_to_income_ratio[i] > 0.5:
        risk += 1

    if repayment_history[i] == "Poor":
        risk += 1

    if credit_score[i] < 500:
        risk += 1

    if risk >= 2:
        loan_status.append(1)
    else:
        loan_status.append(np.random.choice([0, 1], p=[0.7, 0.3]))

loan_status = np.array(loan_status)


df = pd.DataFrame({

    "applicant_id": applicant_id,
    "age": age,
    "gender": gender,
    "region": region,
    "education": education,
    "employment": employment,
    "employment_years": employment_years,
    "monthly_income": monthly_income,
    "loan_amount": loan_amount,
    "loan_duration": loan_duration,
    "interest_rate": interest_rate,
    "loan_purpose": loan_purpose,
    "credit_score": credit_score,
    "dependents": dependents,
    "previous_loans": previous_loans,
    "repayment_history": repayment_history,
    "collateral": collateral,
    "debt_to_income_ratio": debt_to_income_ratio,
    "loan_status": loan_status
})


df.to_csv("microfinance_final_project_dataset.csv", index=False)

print("Synthetic microfinance dataset generated successfully.")
```
